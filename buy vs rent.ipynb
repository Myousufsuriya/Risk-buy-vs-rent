```python
def rent_vs_buy_simple_beginner():

    # --- SIMPLIFIED INPUTS ---
    house_price = float(input("House price (PKR): "))
    loan_years = int(input("Loan tenure in years: "))
    loan_interest_rate = float(input("Annual loan interest rate (%): "))

    initial_monthly_rent = float(input("Initial monthly rent (PKR): "))
    annual_rent_growth = float(input("Annual rent growth (%): "))

    stock_return_annual = float(input("Annual stock market return (%): "))
    house_appreciation_rate = float(input("Annual house appreciation rate (%): ")) # NEW INPUT

    # --- EMI CALCULATION ---
    def calculate_emi(principal, annual_rate, years):
        r_monthly = annual_rate / 100 / 12
        n_months = years * 12
        if r_monthly == 0:
            return principal / n_months if n_months > 0 else 0
        return (principal * r_monthly * (1 + r_monthly) ** n_months) / ((1 + r_monthly) ** n_months - 1)

    monthly_emi = calculate_emi(house_price, loan_interest_rate, loan_years)

    total_housing_interest_paid = (monthly_emi * 12 * loan_years) - house_price
    if total_housing_interest_paid < 0: # Ensure interest is not negative if EMI somehow less than principal over time
        total_housing_interest_paid = 0
    total_principal_paid_housing = house_price

    # --- SIMULATION --- (Simplified)
    stock_rent = 0
    stock_buy = 0
    total_invest_rent = 0
    total_rent_paid = 0

    current_monthly_rent = initial_monthly_rent

    for year in range(loan_years):
        yearly_emi = monthly_emi * 12
        yearly_rent = current_monthly_rent * 12

        # For Rent + Invest: invest the difference (EMI - Rent)
        invest_rent_yearly = max(0, yearly_emi - yearly_rent)
        # For Buy + Invest: no additional investment
        invest_buy_yearly = 0 # As per user's last instruction

        total_invest_rent += invest_rent_yearly
        total_rent_paid += yearly_rent

        stock_rent = (stock_rent + invest_rent_yearly) * (1 + stock_return_annual / 100)
        # stock_buy remains 0 as invest_buy_yearly is 0

        current_monthly_rent *= (1 + annual_rent_growth / 100)

    # Calculate final house value with appreciation
    final_house_value = house_price * (1 + house_appreciation_rate / 100) ** loan_years

    # --- RESULTS ---
    print("\n===== Simplified Rent vs. Buy Results =====")

    print("\n--- Rent + Invest Strategy ---")
    print(f"Total Rent Paid: PKr {total_rent_paid:,.2f}")
    print(f"Total Invested (from EMI-Rent difference): PKr {total_invest_rent:,.2f}")
    print(f"Final Investment Value (Rent Strategy): PKr {stock_rent:,.2f}")

    print("\n--- Buy Strategy ---")
    print(f"Initial House Price: PKr {house_price:,.2f}")
    print(f"Total Principal Paid Towards House: PKr {total_principal_paid_housing:,.2f}")
    print(f"Total Interest Paid Towards House: PKr {total_housing_interest_paid:,.2f}")
    print(f"Final Appreciated House Value: PKr {final_house_value:,.2f}")
    print(f"Net Wealth (Buy Strategy, Appreciated House Value): PKr {final_house_value:,.2f}")

    print("\n--- Overall Comparison ---")
    net_wealth_rent = stock_rent
    net_wealth_buy = final_house_value # Now includes house appreciation

    print(f"Net Wealth (Rent Strategy): PKr {net_wealth_rent:,.2f}")
    print(f"Net Wealth (Buy Strategy): PKr {net_wealth_buy:,.2f}")

    if net_wealth_rent > net_wealth_buy:
        decision = "RENT + INVEST"
    else:
        decision = "BUY"

    print(f"\nFinal Decision: {decision}")

    print("\n📌 Interpretation:")
    if decision == "BUY":
        print("Buying wins because the house's appreciated value surpassed the rental investment value.")
    else:
        print("Renting wins because investing the difference between EMI and rent yielded a higher return in the stock market than the house's appreciated value.")

# Call the new simplified function
rent_vs_buy_simple_beginner()
```
